```python
import requests
import pandas as pd
```

```python
df = pd.read_csv("flowering_genes.csv")
df = df.drop(df.index[:3])
df
```

```python
import re

def fetch_fasta(url):
    gene_id = re.search(r'uniprotkb/(\w+)', url).group(1)
    response = requests.get(f"https://rest.uniprot.org/uniprotkb/{gene_id}.fasta")
    if response.status_code == 200:
        return response.text
    else:
        return None

fasta_data = df["evidence_url"].apply(lambda x: fetch_fasta(x) if pd.notnull(x) else x)
```

```python
fasta_data
```

```python
df["fasta"] = fasta_data
df
```

```python
df.to_csv("flowering_genes_with_fasta.csv", index=False)
```
